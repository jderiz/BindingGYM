```python
model_group = {'ProteinMPNN':'Structure-based','ESM-if1':'Structure-based','PiFold':'Structure-based','ByProt':'Structure-based','PPIformer':'Structure-based','SaProt':'Structure-based',
              'ProGen2':'Protein Language-based','ESM1v':'Protein Language-based','ESM2':'Protein Language-based','ESM3':'Protein Language-based',
              'EVE':'MSA-based','Tranception':'MSA+Protein Language-based','TranceptEVE':'MSA+Protein Language-based'}
```

```python
zero_shot_model_order = ['ProteinMPNN','ESM-if1','PiFold','ByProt','PPIformer','SaProt','ProGen2','ESM1v','ESM2','ESM3','EVE','Tranception','TranceptEVE']
zero_shot_metric = {}
for model in zero_shot_model_order:
    df = pd.read_csv(f'./results/{model}_zero_shot_metric.csv')
    zero_shot_metric[model] = dict(df[['Spearman','AUC']].mean())
    sample_spearman = []
    sample_auc = []
    for _ in range(1000):
        tmp = df.sample(frac=1,replace=True)
        sample_spearman.append(tmp['Spearman'].mean())
        sample_auc.append(tmp['AUC'].mean())
    zero_shot_metric[model]['Spearman Std. error'] = np.std(sample_spearman)
    zero_shot_metric[model]['AUC Std. error'] = np.std(sample_auc)
#         break
```

```python
zero_shot_metric_df = pd.DataFrame(zero_shot_metric.values())
```

```python
zero_shot_metric_df.insert(0,'Model',zero_shot_metric.keys())
zero_shot_metric_df.insert(0,'Category',zero_shot_metric_df['Model'].map(model_group))
```

```python
amino_acids = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', \
               'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y']
mutant_count = np.zeros((20,20))
for DMS_id in bindingGYM['DMS_id'].values:
    print(DMS_id)
    df = pd.read_csv(f'input/Binding_substitutions_DMS/{DMS_id}.csv')
    for i in df.index:
        mutant = eval(df.loc[i,'mutant'])
        for c in mutant:
            if mutant[c] == '':continue
            for m in mutant[c].split(':'):
                i = amino_acids.index(m[0])
                j = amino_acids.index(m[-1])
                mutant_count[i,j] += 1
```

```python
plt.figure(figsize=(6,6))
norm = TwoSlopeNorm(vmin=0.,vcenter=5000, vmax=10000)
plt.imshow(mutant_count, norm=norm, cmap="Blues")
plt.colorbar(fraction=0.04)
plt.ylabel('Wild-type amino acid')

plt.xlabel('Mutant amino acid')
plt.xticks(np.arange(20),amino_acids)
plt.yticks(np.arange(20),amino_acids)
plt.tight_layout()
plt.savefig('./results/mutant_aa_stat.png',dpi=500)
```
